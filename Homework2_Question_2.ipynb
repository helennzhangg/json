```python
#import libraries 
import requests 
import json
import pandas as pd
```

```python
#creating a list for cities the program has to run through 
#creating empty lists for data captured 
citynames = ['New York', 'Tokyo', 'Delhi', 'Shanghai', 'São Paulo', 'Mexico City', 'Cairo', 'Dhaka', 'Mumbai', 'Beijing', 'Osaka']
temperature = []
feelslike = []
humidity = []
mintemp = []
maxtemp = []
```

```python
#creating for loop because code has to be ran through multiple cities 
for city in citynames:
  #my key for the open weather map api
  key = '3e72b86ec3b6e185c0edc1e56e6b63ff'
  #given url for api
  url = f'https://api.openweathermap.org/data/2.5/weather?q={city}&APPID={key}&units=imperial'
  #making the request to API
  r = requests.get(url)
  #retrieving the json sent by API
  data = r.json()
  #print(r.status_code) ; status code is 200
  #using json dumps to return string + make code readable 
  FormatJSON = json.dumps(data, sort_keys = True, indent = 8)
  #print(FormatJSON)
  
  #appending the empty lists
  #there isn't a list in data for main
  temperature.append(data['main']['temp'])
  feelslike.append(data['main']['feels_like'])
  humidity.append(data['main']['humidity'])
  mintemp.append(data['main']['temp_min'])
  maxtemp.append(data['main']['temp_max'])
  
  #create dictionary for upcoming dataframe 
  weather = {
      'Name' : citynames,
      'Temperature' : temperature,
      'Feels Like' : feelslike,
      'Humidity' :humidity,
      'Minimum Temperature' : mintemp,
      'Maximum Temperature' : maxtemp
  }
```

```python
#creating dataframe from dictionary 
weatherreport = pd.DataFrame(weather)
weatherreport
```

```python
#retrieving stats using .describe()
weatherreport.describe()
```

```python
#converting dataframe into csv file named Weather Report
weatherreport.to_csv('Weather Report.csv', index = False)
```
